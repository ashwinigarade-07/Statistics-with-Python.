```python
#Volume Column
Volume_mean=df['Volume'].mean()
print("Volume Mean: ",Volume_mean)
Volume_median=df['Volume'].median()
print("Volume Median: ",Volume_median)
Volume_mode=df['Volume'].mode()[0]
print("Volume mode: ",Volume_mode)
Volume_var=df['Volume'].var()
print("Volume var: ",Volume_var)
Volume_std=df['Volume'].std()
print("Volume std: ",Volume_std)
```

```python
#Avg Price column
AvgPrice_mean = df['Avg Price'].mean()
print("Avg price mean: ",AvgPrice_mean)
AvgPrice_median = df['Avg Price'].median()
print("Avg Price median: ",AvgPrice_median)
AvgPrice_mode = df['Avg Price'].mode()[0]
print("Avg Price mode: ",AvgPrice_mode)
AvgPrice_var =df['Avg Price'].var()
print("Avg Price var: ",AvgPrice_var)
AvgPrice_std =df['Avg Price'].std()
print("Avg Price std: ",AvgPrice_std)
```

```python
#Total Sales Value
TotalSalesValue_mean = df['Total Sales Value'].mean()
print("Total Sales Value mean:",TotalSalesValue_mean)
TotalSalesValue_median = df['Total Sales Value'].median()
print("Total Sales Value median:",TotalSalesValue_median)
TotalSalesValue_mode = df['Total Sales Value'].mode()[0]
print("Total Sales Value mode:",TotalSalesValue_mode)
TotalSalesValue_var = df['Total Sales Value'].var()
print("Total Sales Value var:",TotalSalesValue_var)
TotalSalesValue_std = df['Total Sales Value'].std()
print("Total Sales Value std:",TotalSalesValue_std)
```

```python
#Discount Rate (%)
DiscountRate_mean = df['Discount Rate (%)'].mean()
print("Discount Rate mean:",DiscountRate_mean)
DiscountRate_median = df['Discount Rate (%)'].median()
print("Discount Rate median:",DiscountRate_median)
DiscountRate_mode = df['Discount Rate (%)'].mode()[0]
print("DiscountRate mode:",DiscountRate_mode)
DiscountRate_var = df['Discount Rate (%)'].var()
print("Discount Rate var:",DiscountRate_var)
DiscountRate_std = df['Discount Rate (%)'].std()
print("Discount Rate std:",DiscountRate_std)
```

```python
#Discount Amount
DiscountAmount_mean = df['Discount Amount'].mean()
print("Discount Amount mean:",DiscountAmount_mean)
DiscountAmount_median = df['Discount Amount'].median()
print("Discount Amount Median:",DiscountAmount_median)
DiscountAmount_mode = df['Discount Amount'].mode()[0]
print("Discount Amount Mode",DiscountAmount_mode)
DiscountAmount_var = df['Discount Amount'].var()
print("Discount Amount var:",DiscountAmount_var)
DiscountAmount_std = df['Discount Amount'].std()
print("Discount Amount std:",DiscountAmount_std)
```

```python
#Net Sales Value
NetSalesValue_mean = df['Net Sales Value'].mean()
print("Net Sales Values mean:",NetSalesValue_mean)
NetSalesValue_median = df['Net Sales Value'].median()
print("Net Sales Values median:",NetSalesValue_median)
NetSalesValue_mode = df['Net Sales Value'].mode()[0]
print("Net Sales Values mode:",NetSalesValue_mode)
NetSalesValue_var = df['Net Sales Value'].var()
print("Net Sales Values var:",NetSalesValue_var)
NetSalesValue_std = df['Net Sales Value'].std()
print("Net Sales Values std:",NetSalesValue_std)
```

```python
#mean --> it is comes under measure of central tendency,means is applicable only on numerical data , mean gives the central value of dataset ,if there is no outliers are present then the mean is caluculated
#median--> it is comes under measure of central tendency,median is applicable on numeric data ,while calculating mean first data is sorted in ascending order if there are od numbers of values then middle value is consider as median and if there are even no. of values in dataset then avg of two middle values is taken and if outliers are present then median is caluculated
#mode --> it is comes under measure of central tendency, mode calculates the most frequently occuring value in dataset ,it is applicable on the categorical data
#variance -->it is comes under measure of variance/dispersion , variance is nothing but calculating the spread of data means calculating the distance of data from mean
#standard deviation --> it is comes under measure of variance/dispersion , standard deviation is square root of variance ,low std means --values are close to mean ,high std means-- higly spread of the data
```
