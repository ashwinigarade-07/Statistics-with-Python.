```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Patient_Data.csv")
df
```

```python
df.info() #df.info() -->gives brief description about data that it contains how may rows, not null value , data type of column
```

```python
df = df[['PatientID', 'Department', 'Doctor','BillAmount']] #-->it select only those column from dataframe which is pass inside the square brakcet
df
```

```python
df.drop(columns=['ReceptionistID', 'CheckInTime'], inplace=True, errors='ignore') #drop is use to delete the column from the dataset
```

```python
df
```

```python
df.groupby('Department')['BillAmount'].mean() #groupby first group the deparment of the same name and then calculate the mean of BillAmount as per department
```

```python
df[df.duplicated()] #duplicated is use to find which row is duplicate
```

```python
df.drop_duplicates('PatientID',inplace=True)  #drop_duplicate is use to delete the duplicate row
df
```

```python
df.fillna(df['BillAmount'].mean(),inplace=True)  #fillna method is use to fill missing value in any column by using one of the measure of central tendency
df
```

```python
billing_df= pd.read_csv("Billing_Data.csv") #by using pandas we are reading the data which is present inside Billing_Data.csv
billing_df
```

```python
merge_df = pd.merge(df, billing_df, on='PatientID') #merge method is use to merge two different dataset
merge_df
```
